```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_palette("Set1")
sns.set_context('paper')
```

# Visualize Towers of Hanoi Results

```python
results_path_hanoi = "results/"
#disk_5_q = pd.read_csv(results_path_dqn + "MLP_agent_stats.csv")
disk_5_imi = pd.read_csv(results_path_hanoi + "5_RUNS_5_DISKS_Imitation-SMDP-Q-Learning_TOH.csv")
disk_5_trafo = pd.read_csv(results_path_hanoi + "5_RUNS_5_DISKS_Transfer-SMDP-Q-Learning_1_TOH.csv")
disk_5_imi.columns.values
```

```python
its_imi_5 = disk_5_imi.loc[:, "opt_counter"]
rew_med_imi_5 = disk_5_imi.loc[:, "rew_median"]
rew_10_imi_5 = disk_5_imi.loc[:, "rew_10th_p"]
rew_90_imi_5 = disk_5_imi.loc[:, "rew_90th_p"]

its_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "opt_counter"]
rew_med_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "rew_median"]
rew_10_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "rew_10th_p"]
rew_90_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "rew_90th_p"]
```

```python
its_ticks = np.arange(10000, np.max(its_imi_5), 10000)
its_labels_temp = [str(int(it/10000)) for it in its_ticks]
its_labels = [it_l for it_l in its_labels_temp]
its_labels[0] = r"$1$"
        
plt.plot(figsize=(15,15))
rew_1 = plt.plot(its_imi_5, rew_med_imi_5, label="Expert 5-Disk Grammar")
plt.fill_between(its_imi_5, rew_10_imi_5, rew_90_imi_5, color=rew_1[0].get_color(), alpha=0.25)

rew_2 = plt.plot(its_trafo_5, rew_med_trafo_5, label="Transfer 4-Disk Grammar")
plt.fill_between(its_trafo_5, rew_10_trafo_5, rew_90_trafo_5, color=rew_2[0].get_color(), alpha=0.25)

plt.ylabel("Episode Reward", fontsize=15)
#plt.ylim(bottom=-10, top=450)
plt.xlabel(r"$\times 10^4$ Iterations", fontsize=15)
plt.legend(fontsize=10, loc=4)
plt.title("5 Disk Towers of Hanoi", fontsize=17)
plt.xticks(its_ticks, its_labels, fontsize=9)
#plt.tight_layout()
plt.savefig("figures/hanoi_5_learning_curve.png", dpi=300)
```

# Visualize DQN Results

```python
results_path_dqn = "experiments/results/"
dueling_dqn = pd.read_csv(results_path_dqn + "MLP_agent_stats.csv")
dueling_dqn.columns.values
```

```python
its = dueling_dqn.loc[:150, "opt_counter"]
rew_med = dueling_dqn.loc[:150, "rew_median"]
rew_10 = dueling_dqn.loc[:150, "rew_10th_P"]
rew_90 = dueling_dqn.loc[:150, "rew_90th_p"]
```

```python
from scipy.interpolate import make_interp_spline, BSpline

its_n = np.linspace(its.min(),its.max(), 100)
spl = make_interp_spline(its, rew_med, k=3)
power_smooth = spl(its_n)

plt.plot(its_n, power_smooth)
```

```python
its_ticks = np.arange(1000, np.max(its), 1000)
its_labels_temp = [str(int(it/1000)) for it in its_ticks]
its_labels = [it_l for it_l in its_labels_temp]
its_labels[0] = r"$1$"
        
plt.plot(figsize=(15,15))
rew_p = plt.plot(its, rew_med, label="Dueling DQN")
plt.fill_between(its, rew_10, rew_90, color=rew_p[0].get_color(), alpha=0.5)

plt.ylabel("Episode Reward", fontsize=15)
plt.ylim(bottom=-10, top=450)
plt.xlabel(r"$\times 10^3$ Iterations", fontsize=15)
plt.legend(fontsize=12, loc=4)
plt.title("5-Runs Dueling DQN - Gridworld", fontsize=17)
plt.xticks(its_ticks, its_labels, fontsize=9)
#plt.tight_layout()
plt.savefig("figures/mlp_ddqn_learning_curve.png", dpi=300)
```
