```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_palette("Set1")
sns.set_context('paper')
```

# Visualize Towers of Hanoi Results

```python
results_path_hanoi = "results/"

disk_4_q = pd.read_csv(results_path_hanoi + "5_RUNS_4_DISKS_Q-Learning_TOH.csv")
disk_4_imi = pd.read_csv(results_path_hanoi + "5_RUNS_4_DISKS_Imitation-SMDP-Q-Learning_TOH.csv")

disk_5_q = pd.read_csv(results_path_hanoi + "5_RUNS_5_DISKS_Q-Learning_TOH.csv")
disk_5_imi = pd.read_csv(results_path_hanoi + "5_RUNS_5_DISKS_Imitation-SMDP-Q-Learning_TOH.csv")
disk_5_trafo = pd.read_csv(results_path_hanoi + "5_RUNS_5_DISKS_Transfer-SMDP-Q-Learning_1_TOH.csv")

# disk_6_q = pd.read_csv(results_path_hanoi + "5_RUNS_6_DISKS_Q-Learning_TOH.csv")
disk_6_imi = pd.read_csv(results_path_hanoi + "5_RUNS_6_DISKS_Imitation-SMDP-Q-Learning_TOH.csv")
# disk_6_trafo_1 = pd.read_csv(results_path_hanoi + "5_RUNS_6_DISKS_Transfer-SMDP-Q-Learning_1_TOH.csv")
disk_6_trafo_2 = pd.read_csv(results_path_hanoi + "5_RUNS_6_DISKS_Transfer-SMDP-Q-Learning_2_TOH.csv")

disk_5_imi.columns.values
```

```python
its_imi_4 = disk_4_imi.loc[:, "opt_counter"]
rew_med_imi_4 = disk_4_imi.loc[:, "rew_median"]
rew_10_imi_4 = disk_4_imi.loc[:, "rew_10th_p"]
rew_90_imi_4 = disk_4_imi.loc[:, "rew_90th_p"]

its_q_4 = disk_4_q.loc[:len(its_imi_4), "opt_counter"]
rew_med_q_4 = disk_4_q.loc[:len(its_imi_4), "rew_median"]
rew_10_q_4 = disk_4_q.loc[:len(its_imi_4), "rew_10th_p"]
rew_90_q_4 = disk_4_q.loc[:len(its_imi_4), "rew_90th_p"]
```

```python
its_imi_5 = disk_5_imi.loc[:, "opt_counter"]
rew_med_imi_5 = disk_5_imi.loc[:, "rew_median"]
rew_10_imi_5 = disk_5_imi.loc[:, "rew_10th_p"]
rew_90_imi_5 = disk_5_imi.loc[:, "rew_90th_p"]

its_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "opt_counter"]
rew_med_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "rew_median"]
rew_10_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "rew_10th_p"]
rew_90_trafo_5 = disk_5_trafo.loc[:len(its_imi_5), "rew_90th_p"]

its_q_5 = disk_5_q.loc[:len(its_imi_5), "opt_counter"]
rew_med_q_5 = disk_5_q.loc[:len(its_imi_5), "rew_median"]
rew_10_q_5 = disk_5_q.loc[:len(its_imi_5), "rew_10th_p"]
rew_90_q_5 = disk_5_q.loc[:len(its_imi_5), "rew_90th_p"]
```

```python
its_imi_6 = disk_6_imi.loc[:, "opt_counter"]
rew_med_imi_6 = disk_6_imi.loc[:, "rew_median"]
rew_10_imi_6 = disk_6_imi.loc[:, "rew_10th_p"]
rew_90_imi_6 = disk_6_imi.loc[:, "rew_90th_p"]

#its_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "opt_counter"]
#rew_med_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "rew_median"]
#rew_10_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "rew_10th_p"]
#rew_90_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "rew_90th_p"]

its_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "opt_counter"]
rew_med_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "rew_median"]
rew_10_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "rew_10th_p"]
rew_90_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "rew_90th_p"]
```

# Figure 1 - a - 5 Disk ToH Q-Learn, Imitation/Expert, Transfer-1

```python
its_ticks = np.arange(10000, np.max(its_imi_5), 10000)
its_labels_temp = [str(int(it/10000)) for it in its_ticks]
its_labels = [it_l for it_l in its_labels_temp]
its_labels[0] = r"$1$"
        
plt.plot(figsize=(15,15))
rew_1 = plt.plot(its_imi_5, rew_med_imi_5, label="Expert 5-Disk Grammar")
plt.fill_between(its_imi_5, rew_10_imi_5, rew_90_imi_5, color=rew_1[0].get_color(), alpha=0.25)

rew_2 = plt.plot(its_trafo_5, rew_med_trafo_5, label="Transfer 4-Disk Grammar")
plt.fill_between(its_trafo_5, rew_10_trafo_5, rew_90_trafo_5, color=rew_2[0].get_color(), alpha=0.25)

rew_3 = plt.plot(its_q_5, rew_med_q_5, label=r"TD($\lambda$)")
plt.fill_between(its_q_5, rew_10_q_5, rew_90_q_5, color=rew_3[0].get_color(), alpha=0.25)

plt.ylabel("Episode Reward", fontsize=15)
#plt.ylim(bottom=-10, top=450)
plt.xlabel(r"$\times 10^4$ Iterations", fontsize=15)
plt.legend(fontsize=10, loc=4)
plt.title("Expert & Transfer Grammar: 5 Disks ToH", fontsize=16)
plt.xticks(its_ticks, its_labels, fontsize=9)
#plt.tight_layout()
plt.savefig("figures/hanoi_5_learning_curve.png", dpi=300)
```

# Visualize DQN Results

```python
results_path_dqn = "experiments/results/"
dqn = pd.read_csv(results_path_dqn + "5_RUNS_MLP_agent_stats.csv")
dqn.columns.values
```

```python
its = dqn.loc[:150, "opt_counter"]
rew_med = dqn.loc[:150, "rew_median"]
rew_10 = dqn.loc[:150, "rew_10th_p"]
rew_90 = dqn.loc[:150, "rew_90th_p"]
```

```python
its_ticks = np.arange(1000, np.max(its), 1000)
its_labels_temp = [str(int(it/1000)) for it in its_ticks]
its_labels = [it_l for it_l in its_labels_temp]
its_labels[0] = r"$1$"
        
plt.plot(figsize=(15,15))
rew_p = plt.plot(its, rew_med, label="DQN")
plt.fill_between(its, rew_10, rew_90, color=rew_p[0].get_color(), alpha=0.5)

plt.ylabel("Episode Reward", fontsize=15)
plt.ylim(bottom=-10, top=450)
plt.xlabel(r"$\times 10^3$ Iterations", fontsize=15)
plt.legend(fontsize=12, loc=4)
plt.title("Grammar DQN - Gridworld", fontsize=17)
plt.xticks(its_ticks, its_labels, fontsize=9)
#plt.tight_layout()
plt.savefig("figures/grammar_dqn_gridworld.png", dpi=300)
```
