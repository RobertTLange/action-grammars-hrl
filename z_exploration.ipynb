```python
its_imi_6 = disk_6_imi.loc[:, "opt_counter"]
rew_med_imi_6 = disk_6_imi.loc[:, "rew_median"]
rew_10_imi_6 = disk_6_imi.loc[:, "rew_10th_p"]
rew_90_imi_6 = disk_6_imi.loc[:, "rew_90th_p"]

its_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "opt_counter"]
rew_med_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "rew_median"]
rew_10_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "rew_10th_p"]
rew_90_trafo_1_6 = disk_6_trafo_1.loc[:len(its_imi_6), "rew_90th_p"]

its_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "opt_counter"]
rew_med_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "rew_median"]
rew_10_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "rew_10th_p"]
rew_90_trafo_2_6 = disk_6_trafo_2.loc[:len(its_imi_6), "rew_90th_p"]

its_q_6 = disk_6_q.loc[:len(its_imi_6), "opt_counter"]
rew_med_q_6 = disk_6_q.loc[:len(its_imi_6), "rew_median"]
rew_10_q_6 = disk_6_q.loc[:len(its_imi_6), "rew_10th_p"]
rew_90_q_6 = disk_6_q.loc[:len(its_imi_6), "rew_90th_p"]
```

# Figure 1 - a - 5 Disk ToH Q-Learn, Imitation/Expert, Transfer-1

```python
its_ticks = np.arange(10000, np.max(its_imi_5), 10000)
its_labels_temp = [str(int(it/10000)) for it in its_ticks]
its_labels = [it_l for it_l in its_labels_temp]
its_labels[0] = r"$1$"
        
plt.plot(figsize=(15,15))
rew_1 = plt.plot(its_imi_5, rew_med_imi_5, label="Expert 5-Disk Grammar")
plt.fill_between(its_imi_5, rew_10_imi_5, rew_90_imi_5, color=rew_1[0].get_color(), alpha=0.25)

rew_2 = plt.plot(its_trafo_5, rew_med_trafo_5, label="Transfer 4-Disk Grammar")
plt.fill_between(its_trafo_5, rew_10_trafo_5, rew_90_trafo_5, color=rew_2[0].get_color(), alpha=0.25)

rew_3 = plt.plot(its_q_5, rew_med_q_5, label=r"TD($\lambda$)")
plt.fill_between(its_q_5, rew_10_q_5, rew_90_q_5, color=rew_3[0].get_color(), alpha=0.25)

plt.ylabel("Episode Reward", fontsize=15)
#plt.ylim(bottom=-10, top=450)
plt.xlabel(r"$\times 10^4$ Iterations", fontsize=15)
plt.legend(fontsize=10, loc=4)
plt.title("Expert & Transfer Grammar: 5 Disks ToH", fontsize=16)
plt.xticks(its_ticks, its_labels, fontsize=9)
#plt.tight_layout()
plt.savefig("figures/hanoi_5_learning_curve.png", dpi=300)
```

# Plot 2 - a - Grid World - Expert Grammar

```python
results_path_dqn = "experiments/results/"
dqn = pd.read_csv(results_path_dqn + "5_RUNS_MLP-DQN_MLP_agent_stats.csv")
dueling_dqn = pd.read_csv(results_path_dqn + "5_RUNS_MLP-Dueling-DQN_MLP_agent_stats.csv")

dqn_g_3000_2macros = pd.read_csv(results_path_dqn + "5_RUNS_MLP-DQN_2_expert_grammar_stats_3000.csv")
dqn_g_3000_4macros = pd.read_csv(results_path_dqn + "5_RUNS_MLP-DQN_4_expert_grammar_stats_3000.csv")

dqn.columns.values
```

```python
num_logs = 75
its_dqn = dqn.loc[:num_logs, "opt_counter"]
rew_med_dqn = dqn.loc[:num_logs, "rew_median"]
rew_10_dqn = dqn.loc[:num_logs, "rew_10th_p"]
rew_90_dqn = dqn.loc[:num_logs, "rew_90th_p"]

its_dueling_dqn = dueling_dqn.loc[:num_logs, "opt_counter"]
rew_med_dueling_dqn = dueling_dqn.loc[:num_logs, "rew_median"]
rew_10_dueling_dqn = dueling_dqn.loc[:num_logs, "rew_10th_p"]
rew_90_dueling_dqn = dueling_dqn.loc[:num_logs, "rew_90th_p"]

its_dqn_g_3000_2macros = dqn_g_3000_2macros.loc[:num_logs, "opt_counter"]
rew_med_dqn_g_3000_2macros = dqn_g_3000_2macros.loc[:num_logs, "rew_median"]
rew_10_dqn_g_3000_2macros = dqn_g_3000_2macros.loc[:num_logs, "rew_10th_p"]
rew_90_dqn_g_3000_2macros = dqn_g_3000_2macros.loc[:num_logs, "rew_90th_p"]

its_dqn_g_3000_4macros = dqn_g_3000_4macros.loc[:num_logs, "opt_counter"]
rew_med_dqn_g_3000_4macros = dqn_g_3000_4macros.loc[:num_logs, "rew_median"]
rew_10_dqn_g_3000_4macros = dqn_g_3000_4macros.loc[:num_logs, "rew_10th_p"]
rew_90_dqn_g_3000_4macros = dqn_g_3000_4macros.loc[:num_logs, "rew_90th_p"]
```

```python
its_ticks = np.arange(1000, np.max(its_dqn), 1000)
its_labels_temp = [str(int(it/1000)) for it in its_ticks]
its_labels = [it_l for it_l in its_labels_temp]
its_labels[0] = r"$1$"
        
plt.plot(figsize=(15,15))
rew_2 = plt.plot(its_dqn_g_3000_2macros, rew_med_dqn_g_3000_2macros, label="2-Grammar-DQN")
plt.fill_between(its_dqn_g_3000_2macros, rew_10_dqn_g_3000_2macros,
                 rew_90_dqn_g_3000_2macros, color=rew_2[0].get_color(), alpha=0.25)

rew_4 = plt.plot(its_dqn_g_3000_4macros, rew_med_dqn_g_3000_4macros, label="4-Grammar-DQN")
plt.fill_between(its_dqn_g_3000_4macros, rew_10_dqn_g_3000_4macros,
                 rew_90_dqn_g_3000_4macros, color=rew_4[0].get_color(), alpha=0.25)

rew_1 = plt.plot(its_dqn, rew_med_dqn, label="DQN")
plt.fill_between(its_dqn, rew_10_dqn, rew_90_dqn, color=rew_1[0].get_color(), alpha=0.25)

rew_3 = plt.plot(its_dueling_dqn, rew_med_dueling_dqn, label="Dueling DQN")
plt.fill_between(its_dueling_dqn, rew_10_dueling_dqn, rew_90_dueling_dqn, color=rew_1[0].get_color(), alpha=0.25)


plt.ylabel("Episode Reward", fontsize=15)
plt.ylim(bottom=-10, top=450)
plt.xlabel(r"$\times 10^3$ Iterations", fontsize=15)
plt.legend(fontsize=10, loc=4)
plt.title("Expert Grammar-DQN", fontsize=17)
plt.xticks(its_ticks, its_labels, fontsize=9)
#plt.tight_layout()
plt.savefig("figures/grammar_dqn_gridworld_expert.png", dpi=300)
```
